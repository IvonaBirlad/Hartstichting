```python
#Importing packages

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.svm import LinearSVC
from sklearn.linear_model import Ridge, LinearRegression, Lasso
from sklearn.feature_selection import RFE

from sklearn import linear_model, metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, GridSearchCV, train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import silhouette_score
from sklearn.impute import SimpleImputer

from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler

from sklearn.metrics import average_precision_score, recall_score, precision_score
from sklearn.metrics import f1_score, accuracy_score, classification_report, confusion_matrix

#Setting up matplotlib for plots
%matplotlib inline

# setting random seed
np.random.seed(42)
```

```python
#Helper functions
def get_feature_importance(model, ftrs):
    if model == lasso_grid:
        coefficients = model.best_estimator_.named_steps['model'].coef_
        imp = np.abs(coefficients)
    else:
        imp = model.best_estimator_.feature_importances_.tolist()
    feat = ftrs
    result = pd.DataFrame({'feat':feat,'score':imp})
    result = result.sort_values(by=['score'],ascending=False)
    fig, ax = plt.subplots(1, 1, figsize=(12, 7))
    sns.barplot(x="feat",
                y="score",
                data=result.head(20))
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize=20)
    ax.set_title("Top 20 Features", fontsize=25)
    ax.set_ylabel("Coef", fontsize=22)
    ax.set_xlabel("Feature Name", fontsize=22)
    return result
```

```python
# fixing decimal numbers with comma to proper decimal numbers
features = ["g_hhgro", "g_afs_hp", "g_afs_gs", "g_afs_kv", "g_afs_sc", "g_3km_sc"]

for feature in features:
    merged[feature] = merged[feature].str.replace(",", ".")

merged = merged.astype(float)
```

```python
# making the target the last column
merged = merged[[c for c in merged if c != "Percentage"] + ["Percentage"]]
```

```python
# transforming some of the absolute numbers to percentages
# (e.g. instead of "there are 2347 Westerners in this Buurt" we now know "20% of people int he Buurt are Westerners")
population_parameters = ["a_man", "a_vrouw", "a_00_14", "a_15_24", "a_25_44", "a_45_64",
           "a_65_oo", "a_ongeh", "a_gehuwd", "a_gesch", "a_verwed", "a_w_all",
          "a_nw_all", "a_marok", "a_antaru", "a_suri", "a_tur", "a_ov_nw",
          "a_soz_wb", "a_soz_ao", "a_soz_ww", "a_soz_ow", "a_pau", "a_m2w"]

for feature in population_parameters:
    merged[feature] = merged[feature] / merged["a_inw"]
```

```python
# checking for highly correlated variables and dropping them

X = merged.iloc[:, :-1]
y = merged.iloc[:, -1]

#Finding correlated features
corr_df = X.corr().abs()
to_drop= set()
for i in range(len(corr_df.columns)):
    for j in range(i):
        if corr_df.iloc[i, j] > 0.9:
            colname = corr_df.columns[i]
            to_drop.add(colname)

merged.drop(to_drop, axis=1, inplace=True)
```

## Unsupervised Learning / Clustering

```python
# separating the target from the data
data = merged.iloc[:, :-1]
```

```python
# imputing and scaling the data
imputer = SimpleImputer()
data = imputer.fit_transform(data)

scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

```python
# "Elbow Method" as rule of thumb for chosing the number of clusters --> 2 clusters
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10)
    kmeans.fit(data)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
```

```python
#Clustering with KMeans

km = KMeans(n_clusters=2)
pred = km.fit_predict(data)
print(km.cluster_centers_.shape)
print(km.labels_.shape)
```

```python
for n_clusters in range(2,10):
    km = KMeans(n_clusters=n_clusters)
    pred = km.fit_predict(data)
    print(km.cluster_centers_.shape)
    print(km.labels_.shape)
    print(silhouette_score(data, pred))
```

```python
# adding the classification label to the data
merged["cluster"] = pred

# making the target the last column again
merged = merged[[c for c in merged if c != "Percentage"] + ["Percentage"]]
```

## Lasso Regression

```python
lasso_minmax_pipe = Pipeline([("scaler", MinMaxScaler()), ("model", Lasso(tol=1e-2))])
lasso_standard_pipe = Pipeline([("scaler", StandardScaler()), ("model", Lasso(tol=1e-2))])
```
